```python
def _normalize_commitment(value: Any) -> Dict[str, Any]:
    if isinstance(value, dict):
        data = value.get("data") or value.get("commitment") or value.get("value")
        block = value.get("block") or value.get("block_number") or value.get("submitted_at")
        return {"data": data, "block": block}
    if isinstance(value, (list, tuple)) and len(value) >= 2:
        return {"data": value[0], "block": value[1]}
    return {"data": value, "block": None}


async def get_chain_commitments() -> Dict[int, Dict[str, Any]]:
    def _fetch() -> Dict[int, Dict[str, Any]]:
        subtensor = Subtensor(network=BT_NETWORK)
        raw = subtensor.get_all_commitments(NETUID)
        if isinstance(raw, dict):
            return {int(uid): _normalize_commitment(value) for uid, value in raw.items()}
        if isinstance(raw, list):
            return {uid: _normalize_commitment(value) for uid, value in enumerate(raw)}
        return {}

    return await asyncio.to_thread(_fetch)
```
